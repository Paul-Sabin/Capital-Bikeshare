```python
def categorize_wind_speed(speed):
    if 0 <= speed < 5:
        return 'still'
    elif 5 <= speed < 10:
        return 'waft'
    elif 10 <= speed < 20:
        return 'breeze'
    elif 20 <= speed < 30:
        return 'blow'
    elif 30 <= speed < 40:
        return 'gust'

hour_weather['wind_speed_category'] = hour_weather['wind_speed'].apply(categorize_wind_speed)
```

```python
# Set the size of the figure
plt.figure(figsize=(12, 6))

# Create the lineplot
sns.lineplot(x='hour', y='rides', hue='wind_speed_category', data=hour_weather)

# Set x-axis ticks to display all hours
plt.xticks(range(1, 25))

# Show the plot
plt.show()

```

^ This doesn't really tell us anything, since there were not equal amounts of time at each wind speed, so the number of rides would be different regardless of rider choice based on wind.

```python
# Group by 'hour' and 'wind_speed_category' and calculate the mean for 'rides'
mean_rides_df = hour_weather.groupby(['hour', 'wind_speed_category'])['rides'].mean().reset_index()
display(mean_rides_df.head(20))
```

```python
# Count occurrences
count_df = hour_weather.groupby(['hour', 'wind_speed_category']).count().reset_index()

# Rename the 'rides' column to 'occurrences'
count_df.rename(columns={'rides': 'occurrences'}, inplace=True)

# Merge with mean_rides_df
mean_rides_df = pd.merge(mean_rides_df, count_df[['hour', 'wind_speed_category', 'occurrences']], on=['hour', 'wind_speed_category'])

display(mean_rides_df.head(20))
```

```python
# Set the figure size
plt.figure(figsize=(12, 6))

ordered_categories = ['still', 'waft', 'breeze', 'blow', 'gust']

wind_lines = sns.lineplot(x='hour', y='rides', hue='wind_speed_category', data=mean_rides_df, hue_order=ordered_categories)

# Set the y-axis label
wind_lines.set_ylabel("Mean number of rides")

# Set x-axis ticks to display all hours
plt.xticks(range(1, 25))

# Show the plot
plt.show()
```

^ Now with normalised values. Higher wind speeds make more of a difference outside of peak commuting times. So it appears that leisure rides are more impacted by poor weather than commutes.

```python
# Set the figure size
plt.figure(figsize=(12, 6))

# Create the lineplot and store the AxesSubplot object
day_lines = sns.lineplot(x='hour', y='rides', hue='weekend', palette='hls', data=hour_weather)

# Set the y-axis label
day_lines.set_ylabel("Mean number of rides")

# Get the colors used for the lines in the plot
day_line_colors = [line.get_color() for line in day_lines.lines[:2]]

# Calculate the overall mean number of rides for each category
mean_weekday = np.mean(hour_weather[hour_weather['weekend'] == False]['rides'])
mean_weekend = np.mean(hour_weather[hour_weather['weekend'] == True]['rides'])

# Add horizontal lines for the overall means
plt.axhline(y=mean_weekday, color=day_line_colors[0], linestyle='--', label=f"Overall Mean (Mon-Fri): {mean_weekday:.0f}")
plt.axhline(y=mean_weekend, color=day_line_colors[1], linestyle='--', label=f"Overall Mean (Weekend): {mean_weekend:.0f}")

# Update the legend to include the new lines
handles, labels = day_lines.get_legend_handles_labels()
day_lines.legend(handles=handles, labels=labels, title='Day')

handles, labels = day_lines.get_legend_handles_labels()
labels[0] = 'Mon-Fri'
labels[1] = 'Weekend'
day_lines.legend(handles=handles[0:4], labels=labels[0:4], title='Day')

# Set x-axis ticks to display all hours
plt.xticks(range(1, 25))

# Show the plot
plt.show()
```

^ Overall, the usage per hour is similar on weekends and during the week, which is good for Capital Bikeshare
